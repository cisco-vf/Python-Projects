### Transformación de datos

```python
#Analizamos los distritos de New York y observamos que hay 2 distritos (Manhattan y Brooklyn) que tienen doble registro por estar mal redactados
airbnb['neighbourhood_group'].value_counts()
```

```python
#Corregimos los distritos mal redactados
airbnb['neighbourhood_group'].replace(['manhatan','brookln'],['Manhattan','Brooklyn'],inplace=True)
airbnb['neighbourhood_group'].value_counts()
```

```python
#ServiceFee y Price traen símbolo $, se lo quitaremos para poder cambiar el tipo de dato de object a float
airbnb['price']=airbnb['price'].apply(lambda x: x.replace('$','') if pd.isna(x) == False else x)
airbnb['service_fee']=airbnb['service_fee'].apply(lambda x: x.replace('$','') if pd.isna(x) == False else x)

#Cambiamos el tipo de dato de price y ServiceFee
airbnb['service_fee']=airbnb['service_fee'].astype(float)
airbnb['price']=airbnb['price'].apply(lambda x: x.replace(',','') if pd.isna(x) == False else x)
airbnb['price']=airbnb['price'].astype(float)
```

```python
#Last_Review está en tipo dato object, lo pasaremos a DateTime para poder usarlo como fecha
airbnb['last_review'] = pd.to_datetime(airbnb['last_review'])

#Last_Review trae dato con año: 2058 -> esto es un dato erróneo, y genera gran dispersión en los datos.
#Por lo tanto lo reemplazaremos por la media.
airbnb.loc[airbnb[airbnb['last_review'].apply(lambda x: x.year) > 2022].index,'last_review'] = airbnb['last_review'].mean()

#Last_Review reemplazaremos los valores null por la media también
airbnb.loc[airbnb['last_review'].isnull(), 'last_review'] = airbnb['last_review'].mean()

#Revisamos que haya quedado correctamente Last_Review
airbnb['last_review'].min(), airbnb['last_review'].max()
```

```python
#Cambiamos los valores nulos de la variable 'Construction_year' por la media
#Asimismo, cambiamos el tipo de variable 'Construction_year' de Float a Integer
airbnb.loc[airbnb['Construction_year'].isnull(), 'Construction_year'] = airbnb['Construction_year'].mean()
airbnb['Construction_year'] = airbnb['Construction_year'].round().astype(int)
```

```python
#Mostramos el resultado final para la variable 'Construction_year'
airbnb['Construction_year']
```

```python
#minimun_nights trae dato negativo alto: -1223 -> esto es un dato erróneo, y genera gran dispersión en los datos.
#Lo reemplazaremos por la media.
airbnb.loc[airbnb[airbnb['minimum_nights'].apply(lambda x: x) < 0].index,'minimum_nights'] = airbnb['minimum_nights'].mean()
```

```python
#Llenamos los campos con datos nulos en la columna "price" con el número 0
airbnb["price"].fillna(0)
```

```python
#Revisamos datos de las columnas host identity verified, instant_bookable, roomtype, cancellation_policy, review_rate_number y
#lo cambiaremos a tipo de dato categoría
print(airbnb['host_identity_verified'].value_counts())
print(airbnb['instant_bookable'].value_counts())
print(airbnb['room_type'].value_counts())
print(airbnb['cancellation_policy'].value_counts())
print(airbnb['review_rate_number'].value_counts())

#Hacemos el cambio de tipo de dato a Categoría
airbnb = airbnb.astype({'room_type':'category','cancellation_policy':'category',
                        'instant_bookable':'category','host_identity_verified':'category',
                        'review_rate_number':'category','neighbourhood_group':'category'})
```

### Posteriormente procedo a realizar Modelos Estadísticos que nos permitan hacer mayor análisis a las variables con alto nivel de correlación y obtener respuestas que complementen las hipótesis respondidas

## Modelo de Regresión Lineal

```python
#Importamos las librerías necesarias para los Modelos
import sklearn as sk

from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#Definimos las Features numéricas
df_num = airbnb.select_dtypes('number')
df_num['y'] = airbnb['price']
df_num = df_num.drop('price', axis=1)
df_num.head()
```

```python
len(df_num)
```

***Valores duplicados***

```python
#Eliminamos valores duplicados
df_num.drop_duplicates(inplace=True) #df_num = df_num.drop_duplicates()
```

```python
len(df_num)
```

```python
#Nos quedamos únicamente con las columnas:
columnas = df_num.columns.tolist()
columnas
```

```python
df_num['y']
```

**ii) Elegimos un Algoritmo de Regresión para entrenar con los datos elegidos**

```python
# separamos las caracteristicas/features del target
x = df_num[columnas].drop('y', axis=1)
x = x.filter(items=['Construction_year',
 'service_fee',
 'minimum_nights',
 'number_of_reviews',
 'reviews_per_month',
 'calculated_host_listings_count',
 'availability_365'])
```

```python
#Reemplazaremos las columnas de x por la media
for var_med in x:
    x[var_med].value_counts()
    x.loc[x[var_med].isnull(), var_med] = x[var_med].mean()
    x[var_med]=x[var_med].round().astype(int)
```

```python
df_num['y'].value_counts()
df_num.loc[df_num['y'].isnull(), 'y'] = df_num['y'].mean()
y = df_num['y'].round().astype(int)
```

```python
x.info()
```

```python
y.info()
```

```python
# train test split
x_train, x_test, y_train, y_test = train_test_split(x, y,
                                                    test_size=0.2, 
                                                    random_state=42)

print('Tamaño del set de entrenamiento:', x_train.shape)
print('Tamaño del set de testeo:', x_test.shape)
```

```python
print("x_train")
x_train
```

```python
print("y_train")
y_train
```

```python
#Hacemos Regresión
regr = linear_model.LinearRegression()
regr.fit(x_train,y_train)
y_pred = regr.predict(x_test)
y_pred
```

```python
y_test
```

```python
#Imprimimos resultados de Test y Train a analizar
print('Training score: ', regr.score(x_train, y_train))
print('Test score: ', regr.score(x_test,y_test))
```

```python
#Realizamos tabla de predicción del precio sobre el precio original
predictions = pd.DataFrame({"original_Price": y_test.values, "predicted_Price": y_pred})
predictions.head(10)
```

```python
#Graficamos la predicción en mención
fig, ax = plt.subplots(figsize=(10,8))
ax = sns.histplot(predictions, element="poly")
plt.show()
```

**iii) Calculo de Métricas para validar el Modelo**

Evaluación del modelo:

El último paso es evaluar el rendimiento del algoritmo. Este paso es particularmente importante para comparar qué tan bien funcionan los diferentes algoritmos en un conjunto de datos en particular. Para los algoritmos de regresión, se utilizan comúnmente tres métricas de evaluación:

El error absoluto medio (MAE)
El error cuadrático medio (MSE)
Root Mean Squared Error (RMSE)

Asimismo, añadimos el R2 que nos indica una cifra cercana a 1.

```python
score = sk.metrics.r2_score(y_test, y_pred)
mae = sk.metrics.mean_absolute_error(y_test, y_pred)
mse = sk.metrics.mean_squared_error(y_test, y_pred)

print('Mean Absolute Error is: {}'.format(mae))
print('Mean Squared Error is: {}'.format(mse))
print('Root Mean_squared_error is:', np.sqrt(mse))

print('sklearn: R2 score for Linear Regression is: {}'.format(score))
```

```python
#NORMALIZACION
#Aplicamos una transformación de los datos para poder aplicar las propiedades de la distribución normal
from sklearn.preprocessing import StandardScaler
X_std = StandardScaler().fit_transform(X)
```

```python
# Calculo la matriz de covarianza
print('NumPy covariance matrix: \n%s' %np.cov(X_std.T))
```

```python
#Calculo los autovalores y autovectores de la matriz
cov_mat = np.cov(X_std.T)
eig_vals, eig_vecs = np.linalg.eig(cov_mat)

print('Eigenvectors \n%s' %eig_vecs)
print('\nEigenvalues \n%s' %eig_vals)
```

```python
#  Hacemos una lista de parejas (autovector, autovalor) 
eig_pairs = [(np.abs(eig_vals[i]), eig_vecs[:,i]) for i in range(len(eig_vals))]

# Ordenamos estas parejas den orden descendiente con la función sort
eig_pairs.sort(key=lambda x: x[0], reverse=True)

# Visualizamos la lista de autovalores en orden desdenciente
print('Autovalores en orden descendiente:')
for i in eig_pairs:
    print(i[0])
```

#### Se entiende que una vez ordenados los autovalores, que son una medida de la varianza de los datos, la cuestión es decidir, cuál es el menor número de autovectores o componentes principales, con el que podemos expresar “la esencia principal” de la información contenida en ese dataset. Para ello, usaremos un métrica que se conoce como “varianza explicada”, que muestra cuánta varianza se puede atribuir a cada una de estas componentes principales.

```python
# A partir de los autovalores, calculamos la varianza explicada
tot = sum(eig_vals)
var_exp = [(i / tot)*100 for i in sorted(eig_vals, reverse=True)]
cum_var_exp = np.cumsum(var_exp)
```
